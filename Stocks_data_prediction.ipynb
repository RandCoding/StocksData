```python
def get_ticker_stats(ticker):
    try:
        return get_stats(ticker)
    
    except ValueError:
        return None
        
```

```python
for ticker in tickersList:
    print(get_ticker_stats(ticker))
```

```python
# few_days = get_data('FBIO' , start_date = '01/01/1999' , end_date = '01/10/1999')
# few_days

tickers = {}
for ticker in tickersList:
    interval = get_max_available_date(ticker)
    dates=[]
    for date in interval:
        dates.append(datetime.datetime.strptime(date, '%m/%d/%Y').strftime('%d/%m/%y'))
        
    tickers[ticker]= dates
    
```

```python
data=[]
for key, value in tickers.items():
    data.append(get_data(key , start_date = value[0] , end_date = value[1]))
```
